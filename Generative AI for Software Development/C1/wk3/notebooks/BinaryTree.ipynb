```python
import threading
import logging
import time

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

class TreeNode:
    def __init__(self, key):
        """
        Initialize a tree node.

        :param key: The value to be stored in the node. Must be an integer.
        """
        self.left = None
        self.right = None
        self.val = key

class BinaryTree:
    def __init__(self):
        """
        Initialize an empty binary tree.
        """
        self.root = None
        self.lock = threading.Lock()

    def insert(self, key):
        """
        Insert a new key into the binary tree.

        :param key: The value to insert. Must be an integer.
        :raises ValueError: If the key is not an integer.
        """
        if not isinstance(key, int):
            raise ValueError("Only integer keys are allowed")

        with self.lock:
            start_time = time.time()
            if self.root is None:
                self.root = TreeNode(key)
            else:
                self._insert(self.root, key)
            end_time = time.time()

            logging.info(f"Inserted key {key} in {end_time - start_time:.6f} seconds")

    def _insert(self, node, key):
        """
        Helper method to recursively insert a new key into the tree.

        :param node: The current node in the tree.
        :param key: The value to insert.
        """
        if key < node.val:
            if node.left is None:
                node.left = TreeNode(key)
            else:
                self._insert(node.left, key)
        else:
            if node.right is None:
                node.right = TreeNode(key)
            else:
                self._insert(node.right, key)

    def inorder(self, node):
        """
        Perform an in-order traversal of the tree and print the node values.

        :param node: The current node in the tree.
        """
        if node:
            self.inorder(node.left)
            print(node.val, end=' ')
            self.inorder(node.right)

    def search(self, key):
        """
        Search for a key in the binary tree.

        :param key: The value to search for.
        :return: True if the key is found, False otherwise.
        """
        with self.lock:
            start_time = time.time()
            found = self._search(self.root, key)
            end_time = time.time()

            logging.info(f"Searched for key {key} in {end_time - start_time:.6f} seconds")
            return found

    def _search(self, node, key):
        """
        Helper method to recursively search for a key in the tree.

        :param node: The current node in the tree.
        :param key: The value to search for.
        :return: True if the key is found, False otherwise.
        """
        if node is None:
            return False
        if node.val == key:
            return True
        elif key < node.val:
            return self._search(node.left, key)
        else:
            return self._search(node.right, key)

# Example usage
bt = BinaryTree()
bt.insert(8)
bt.insert(3)
bt.insert(10)
bt.insert(1)
bt.insert(6)
bt.insert(4)
bt.insert(7)

print("Inorder traversal of the binary tree:")
bt.inorder(bt.root)

# Example search
print("\nSearch for 6 in the binary tree:")
print("Found" if bt.search(6) else "Not found")
```
